## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=penguins,
    x='bill_length_mm',
    y='bill_depth_mm',
    hue='species',
    palette='deep'
)
plt.title("Bill Length vs. Bill Depth in penguins")
plt.xlabel("Bill length in mm")
plt.ylabel("Bill depth mm")
plt.show()
```

**Your interpretation (2–3 sentences):**

It seems that where the Adelie and Gentoo penguins value Bill depth and Bill length respectively, the Chinstrap penguin seems to value both at once. Perhaps each species uses their unique traits to their advantage? These penguins had to evolve in certain ways for a reason. Hopefully.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
plt.figure(figsize=(8,6))
sns.histplot(penguins['bill_length_mm'], bins=10, color='skyblue')
plt.title("Histogram of Bill Length")
plt.xlabel("Bill length in mm")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(penguins['bill_length_mm'], bins=30, color='salmon')
plt.title("Histogram of Bill Length (30 bins)")
plt.xlabel("Bill Length in mm")
plt.ylabel("Count")
plt.show()
```

**Your comparison (1–2 sentences):**

Using three times the bins, a lot more nuance is visable. In the ten bin graph, it wasn't clear that there were dips in the middle whereas the thirty bin graph makes that obvious.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
plt.figure(figsize=(8,6))
sns.boxplot(
    data=penguins,
    x='species',
    y='body_mass_g',
    palette='pastel'
)
plt.title("Body Mass by Penguin Species")
plt.xlabel("Species")
plt.ylabel("Body Mass in g")
plt.show()
```

**Your summary (1–2 sentences):**

The Adelie and Chinstrap penguins both have very similar distributions of body mass. The Gentoo generally have much higher body mass than either of the other two penguins. The Chinstrap species also have outliers at either two sides of the spectrum, though what that may suggest is hard to say.
